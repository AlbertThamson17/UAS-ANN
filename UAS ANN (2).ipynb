```python
#fit standard scaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
#transform scaled data
dftrans = scaler.transform(df)

dftrans
```

disini kita tentukan mau berapa dimensi pada n_components

```python
#making a pca component
pca = PCA(n_components=2)
```

```python
#fit pca
pca.fit(dftrans)
```

```python
#transform pca shape
pcashape = pca.transform(dftrans)
```

```python
#print shape before pca
print("Shape1",dftrans.shape)
```

```python
#print shape after pca
print("Shape2",pcashape.shape)
```

bentuk data telah di buat dari 4 menjadi 2 dimensi

Fungsi dari PCA ini adalah untuk menyisakan data dengan komponen komponen yang memiliki varian yang tinggi saja. dengan mereduksi data data ini kita dapat membuat mengurangi dimensi data set yang besar tanpa harus menghilangkan data data penting yang memiliki varian yang tinggi

```python
pcagraph = pca.inverse_transform(pcashape)
plt.scatter(dftrans[:, 0], pcashape[:, 1], alpha=0.2)
plt.scatter(pcagraph[:, 0], pcagraph[:, 1], alpha=0.8)
plt.axis('equal');
```

from my prespective, there are no difference between the split of data from 4 dimension to 2 dimension. as you can see the data still looking as being split into 2 type of datas
